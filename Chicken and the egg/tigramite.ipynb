```python
import pandas as pd
import numpy as np
import numpy as np
from tigramite import data_processing as pp
from tigramite.independence_tests import parcorr
from tigramite.pcmci import PCMCI
from tigramite import plotting as tp


file_path = r'C:\Users\anari\OneDrive - INESC TEC\Documentos\Apresentações\Exemplos código\ChickEgg.csv'
df = pd.read_csv(file_path)

df = df.drop(columns=['Unnamed: 0'])
```

```python
var_names = ["chicken", "egg"]
data_array = df.to_numpy()
data_array = np.array(data_array, dtype=np.float64)
dataframe = pp.DataFrame(data_array, 
                         datatime = np.arange(len(df)), 
                         var_names=var_names)
```

```python
ParCorr = parcorr.ParCorr()
pcmci = PCMCI(
    dataframe=dataframe,
    cond_ind_test=ParCorr,
    verbosity=1)
tau_max=2
results = pcmci.run_pcmci(tau_max=tau_max, pc_alpha=0.01)
```

```python
tp.plot_graph(
    val_matrix=results['val_matrix'],
    graph=results['graph'],
    var_names=var_names,
    link_colorbar_label='cross-MCI',
    node_colorbar_label='auto-MCI',
    )
```

```python
tp.plot_time_series_graph(
    figsize=(6, 4),
    val_matrix=results['val_matrix'],
    graph=results['graph'],
    var_names=var_names,
    link_colorbar_label='MCI',
    )
```
